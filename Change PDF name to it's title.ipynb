```python
def process_title(newName):
        
    # Manage the exceptions
    if newName == None:
        newName = 'error'+fileName
        
    if newName == '':
        newName = 'error'+fileName
        
    if newName == 'untitled':
        newName = 'error'+fileName
        
    if newName in list_pdf_titles:
        newName = 'error'+fileName
    
    # Remove the punctations
    newName = ''.join(ch for ch in newName if ch not in exclude)
    newName = newName + '.pdf'
        
    return newName
```
