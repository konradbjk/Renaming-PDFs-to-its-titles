```python
import os
from PyPDF2 import PdfFileReader
import string
import numpy as np
from pprint import pprint
import pandas as pd
```

```python
path = "rename/"

exclude = set(string.punctuation)

list3 = []
list_of_pdf_titles = []
```

```python
list_of_files = []

for fileName in os.listdir(path):
    #fileName = ''.join(ch for ch in fullName if ch not in exclude)
    fullName = os.path.join(path, fileName)
    if (not os.path.isfile(fullName) or fileName[-4:] != '.pdf'):
        print("It is not pdf: {0}".format(fileName))
        continue
    list_of_files.append(fullName)
```

```python
counter = 0
for e in list_of_files:
    counter+=1

print("Total elements: {0}".format(counter))
#pprint(list_1)
```

```python
def get_pdf_title(row):
    
    pdf_file_path = row['fileName']
    pdf_reader = PdfFileReader(open(pdf_file_path, "rb")) 
    if pdf_reader.isEncrypted:
        pdf_reader.decrypt('')
        if pdf_reader.isEncrypted:
            print(pdf_file_path)
        
    title = process_title(pdf_reader.getDocumentInfo().title)
    list_of_pdf_titles.append(title)
    
    return title
```

```python
def process_title(newName):
        
    # Manage the exceptions
    if newName == None:
        newName = 'error'
        
    if newName == '':
        newName = 'error'
        
    if newName == 'untitled':
        newName = 'error'
        
    if newName in list_of_pdf_titles:
        newName = 'error'+ newName
    
    # Remove the punctations
    newName = ''.join(ch for ch in newName if ch not in exclude)
    newName = newName + '.pdf'
        
    return newName
```

```python
def renameFileToPDFTitle(path, row):
    
    newName = get_pdf_title(row)
        
    # Create path with new name
    newFullName = os.path.join(path, newName)
    
    # Try to rename
    try:
        os.rename(row['fileName'], newFullName)
    except IOError as err:
        print("OS error: {0}".format(err))
    except:
        print("Unexpected error:") #sys.exc_info()[0])
```

```python
df_1 = pd.DataFrame(data=list_of_files, columns = ['fileName'])
df_1
```

```python
for index, row in df_1.iterrows():
    if row['title'] != 'error.pdf':
        renameFileToPDFTitle(path, row)
```
